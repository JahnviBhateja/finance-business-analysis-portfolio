```python
import pandas as pd
```

```python
gdp = pd.read_csv("gdp.csv")
inflation = pd.read_csv("inflation.csv")
exchange= pd.read_csv("exchange_rate.csv")
ip= pd.read_csv("industrial_production.csv")
forex = pd.read_csv("forex_reserves.csv")
exports = pd.read_csv("exports.csv")
```

```python
exports = exports.drop(columns=["Unnamed: 0"])
```

```python
forex = forex.drop(columns=["Unnamed: 0"])
```

```python
master = gdp.merge(inflation, on="Year", how="outer")
master = master.merge(exchange, on="Year", how="outer")
master = master.merge(ip, on="Year", how="outer")
master = master.merge(forex, on="Year", how="outer")
master = master.merge(exports, on="Year", how="outer")
master
```

```python
master = master.drop(columns=["Unnamed: 0"])
```
