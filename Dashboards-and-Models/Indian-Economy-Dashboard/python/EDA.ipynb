```python
# Step 1: Import the required libraries and load the master dataset.
import pandas as pd
import matplotlib.pyplot as plt

master = pd.read_csv("Indian_Economy_Master.csv")
master.head()
```

The dataset contains yearly macroeconomic indicators for India. Variables include GDP, Inflation, Exchange Rate, Industrial Production, Forex Reserves, and Exports.

```python
# Step 4: Visualize trends in key macroeconomic indicators over time.

plt.figure(figsize=(10,5))
plt.plot(master["Year"],master["GDP"])
plt.title("GDP over Time")
plt.xlabel("Year")
plt.ylabel("GDP")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(master["Year"],master["Inflation"])
plt.title("Inflation over Time")
plt.xlabel("Year")
plt.ylabel("Inflation")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(master["Year"],master["Exchange rate"])
plt.title("Exchange Rate over Time")
plt.xlabel("Year")
plt.ylabel("Exchange Rate")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(master["Year"],master["Industrial_Production"])
plt.title("Industrial Production over Time")
plt.xlabel("Year")
plt.ylabel("Industrial Production")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(master["Year"], master["Forex_Reserves"])
plt.title("Forex Reserves over Time")
plt.xlabel("Year")
plt.ylabel("Forex Reserves")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(master["Year"], master["Exports"])
plt.title("Exports over Time")
plt.xlabel("Year")
plt.ylabel("Exports")
plt.grid(True)
plt.show()
```

```python
# Step 5: Examine relationships between variables using a correlation matrix.
master.corr(numeric_only=True)
```

```python
corr= master.corr(numeric_only=True)

plt.figure(figsize=(10,5))
plt.imshow(corr)
plt.xticks(range(len(corr.columns)), corr.columns, rotation=45)
plt.yticks(range(len(corr.columns)), corr.columns)
plt.colorbar()
plt.show()
```

```python
# Step 6: Understand the distribution of each economic indicator.
master.hist(figsize=(12,8))
plt.show()
```

```python
master.isna().sum()
```

```python
#Growth rate
master["GDP_Growth"]= master["GDP"].pct_change()*100
```

```python
#growth growth
master["Export_Growth"]= master ["Exports"].pct_change()*100
```

```python
# Step 8: Compare GDP growth with Export growth over time.
plt.figure(figsize=(12,6))

plt.plot(master["Year"], master["GDP_Growth"], label="GDP Growth, linewidth=2")
plt.plot(master["Year"], master["Export_Growth"], label="Export Growth", linewidth=2)
plt.title("GDP vs Export Growth Rate")
plt.xlabel("Year")
plt.ylabel("Growth Rate(%)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Step 9: Detect potential outliers using boxplots.
master.boxplot(figsize=(10,6))
plt.xticks(rotation=45)
plt.show()
```

## Conclusion

The exploratory data analysis reveals that India's macroeconomic indicators have generally exhibited an upward trend over the study period. GDP, exports, industrial production, and foreign exchange reserves show sustained long-term growth, while inflation and exchange rates display greater fluctuations. Correlation analysis indicates positive relationships among most macroeconomic indicators, providing insights into India's economic development over time.
