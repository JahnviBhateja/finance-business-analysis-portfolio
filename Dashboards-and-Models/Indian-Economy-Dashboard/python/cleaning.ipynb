```python
# Step 1: Import the Pandas library for data manipulation.
import pandas as pd
```

```python
# Step 2: Load the Exchange Rate dataset into a DataFrame.
df = pd.read_excel("Exchange rate, new LCU per USD extended backward, period average.xlsx")
```

```python
# Step 3: Inspect the dataset to understand its structure and identify relevant columns.
df.head()
```

```python
df.head(
)
```

```python
# Step 4: Rename the year column for consistency across all datasets.
df = df.rename(columns={"Unnamed: 0": "Year"})
```

```python
df.head()
```

```python
# Step 5: Keep only the Year and India columns required for analysis.
df= df[["Year", "India"]]
```

```python
# Step 6: Rename the indicator column to a meaningful variable name.
df= df.rename(columns={"India": "Exchange rate"})
```

```python
df.head()
```

```python
df.info()
```

```python
# Step 7: Remove unnecessary rows and convert the Year column to integer format.
df= df.drop(0)
```

```python
df.head()
```

```python
df["Year"] = df["Year"].astype(int)
df.head()
```

```python
# Step 8: Save the cleaned dataset for later merging.
import os

os.makedirs("data/cleaned", exist_ok=True)

df.to_csv("data/cleaned/exchange_rate.csv", index= False)
```
